```python
# REF:[https://stackoverflow.com/a/45884249]
def substation_plot(x):
    for i,col in enumerate(list(x.columns)):
        plt.subplot(2,2,i+1)
        plt.title(col)
        plt.hist([x[x.iloc[:,0]==0][col], x[x.iloc[:,0]==1][col]], bins=15)
        plt.legend([0,1], loc='center')

    plt.tight_layout(pad=0.4)
    plt.show()
```
